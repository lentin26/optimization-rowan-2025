```python
import numpy as np
import matplotlib.pyplot as plt


p = np.array([1, 2.5, 4.8, 6.9, 8.2])
t = np.array([0, 2, 4, 6, 8])
trace = []

s = 1e-3
x = [1, 1, 1]  # a, b, c
i = 0
while i < 1e1:

    grad = -np.array([
        (1 + x[1]*np.exp(-x[2]*t))**-1,
        -x[0]*(1 + x[1]*np.exp(-x[2]*t)**-2),
        x[0]*x[1]*t*np.exp(-x[2]*t)*(1 + x[1]*np.exp(-x[2]*t))**-2 
    ]).sum(axis=1)

    x += -s*grad
    loss = ((p - x[0] / (1 + x[1]*np.exp(-x[2]*t)))**2).sum()
    trace.append(loss)
    i += 1

print(x)
plt.plot(trace)
plt.show()
```
